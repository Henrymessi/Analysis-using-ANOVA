```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib
```

```python
all_data = pd.read_csv("Desktop/Project/2021 Project/UI/Xtiana/Tiana_Data.csv")
all_data.head(5)
```

```python
# Renaming the heading where:
# Week_1 = Initial
# Week_2 = Final

all_data.rename(columns={'Week_1_Mean_Weight': 'Initial_Weight',
                     'Week_12_Mean_Weight': 'Final_Weight',
                     'Week_1_Mean_Length': 'Initial_Length',
                     'Week_12_Mean_Length': 'Final_Length'}, inplace = True)
all_data.head(5)
```

```python
# Drop all rows that have NaN/None Values

data = all_data.dropna()
data
```

```python
# extracting the initial and final values of the diets
new_data = data.filter(['Diets', 'Initial_Weight', 'Final_Weight', 'Initial_Length', 'Final_Length'])
new_data.head(5)
```

```python
#Describe data by group 
new_data.groupby('Diets').describe()
```

```python
#Create a boxplot for the Mean_Weight

new_data.boxplot('Final_Weight', by='Diets', figsize=(12, 8))

Diet_A = new_data['Final_Weight'][new_data.Diets == 'Diet_A']

grps = pd.unique(new_data.Diets.values)
d_data = {grp:new_data['Final_Weight'][new_data.Diets == grp] for grp in grps}

k = len(pd.unique(new_data.Diets))  # number of conditions
N = len(new_data.values)  # conditions times participants
n = new_data.groupby('Diets').size()[0] #Participants in each condition
```

```python
# ANOVA analysis for mean_weight using Statsmodels

# Import python packages

import statsmodels.api as sm
from statsmodels.formula.api import ols

# Set up the ANOVA model
mod = ols('Final_Weight ~ Diets', data = new_data).fit()

# Perform the ANOVA analysis
Weight_ANOVA = sm.stats.anova_lm(mod, typ=2)
print(Weight_ANOVA)
```

```python
pair_t = mod.t_test_pairwise('Diets')
pair_t.result_frame
```
